```python
#2.1.b Part 2: Data Cleaning Handling Missing Values: 
    #For numerical columns with missing values, replace them with the column mean
stores['size_sqft'] = stores['size_sqft'].fillna(stores['size_sqft'].mean())
stores.info()
```

```python
#2.1.c For categorical columns with missing values, replace them with the most frequent value
sales['payment_method'] = sales['payment_method'].fillna(sales['payment_method'].mode()[0])
sales.info()
```

```python
#2.1.d For date columns with missing values, use forward fill or backward fill as appropriate
stores['opening_date'] = stores['opening_date'].ffill()
stores.info()
```
